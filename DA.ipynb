```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
raw_data.info()
```

```python
raw_data.describe()
```

```python
print(raw_data["Branch"].unique())
print(raw_data["Branch"].value_counts(normalize=True))
temp = raw_data["Branch"].value_counts(normalize=True)
fig, ax = plt.subplots(figsize=(9,9))
plt.pie(temp,labels=temp.values)
plt.legend(temp.index)
plt.show()
```

```python
print(raw_data["City"].unique())
print(raw_data["City"].value_counts(normalize=True))
temp = raw_data["City"].value_counts(normalize=True)
fig, ax = plt.subplots(figsize=(10, 10))
plt.pie(temp,labels=temp.values)
plt.legend(temp.index)
plt.show()
```

```python
print(raw_data["Customer type"].unique())
print(raw_data["Customer type"].value_counts(normalize=True))
temp = raw_data["Customer type"].value_counts(normalize=True)
plt.pie(temp,labels=temp.values)
plt.legend(temp.index)
plt.show()
```

```python
print(raw_data["Gender"].unique())
print(raw_data["Gender"].value_counts(normalize=True))
temp = raw_data["Gender"].value_counts(normalize=True)
plt.pie(temp,labels=temp.values)
plt.legend(temp.index)
plt.show()
```
